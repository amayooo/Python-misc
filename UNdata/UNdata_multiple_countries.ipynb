```python
import Ipynb_importer
import UNdata_country_profile as UNdata
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
index_url = "http://data.un.org/en/index.html"
response = requests.get(index_url)
soup = BeautifulSoup(response.content,'lxml')
```

```python
urls = ["http://data.un.org/en/"+u['href'] for u in soup.section.find_all('a', href=True)]
countries = [c.previousSibling for c in soup.section.find_all("br")]
print(len(urls) == len(countries))
```

```python
total = dict()
for i in range(len(urls)):
    profile = UNdata.get_profile(urls[i])
    total[countries[i]] = profile
    print("Finished:" + countries[i] + "(" + str(i+1) + "/"+ str(len(urls))+")")
```

```python
for c in countries:
    total[c][1] = total[c][1][['Index', 'Unit', 2019]] # get 2019 indicators
    total[c][0].columns = ['Index', c]
    total[c][1].columns = ['Index', 'Unit', c]
    for i in range(len(total[c][0]['Index'].tolist())):
        if total[c][0]['Index'][i] == "Capital city pop.(000,2019)":
            total[c][0]['Index'][i]= "Capital city pop.(000, 2019)"
```

```python
# compile the general information table
gi = [total[c][0] for c in countries]
GI = gi[0]
for i in range(len(gi))[1:]:
    GI = pd.merge(GI, gi[i], how="outer", on="Index")
    
# transpose df
GIt = GI.transpose()[1:]
GIt.columns = GI.Index

# convert formats
GIt.loc[:,"Capital city pop.(000, 2019)"] = pd.to_numeric(GIt.loc[:,"Capital city pop.(000, 2019)"],errors="coerce")
GIt.loc[:,"Surface area(km2)"] = pd.to_numeric(GIt.loc[:,"Surface area(km2)"],errors="coerce")
GIt.loc[:,"UN membership date"] = pd.to_datetime(GIt.loc[:,"UN membership date"], infer_datetime_format=True)
```

```python
# Compile the indicators table
ind = [total[c][1] for c in countries]
Ind = ind[0]
for i in range(len(ind))[1:]:
    Ind = pd.merge(Ind, ind[i], how="outer", on= ["Index", "Unit"])

# transpose df
Indt = Ind.transpose()[2:]
Indt.columns = Ind.Index + Ind.Unit

# convert formats
ls = ['Unnamed: 0',
      'Labour force participation(female/male pop. %)',
      'Life expectancy at birth(females/males, years)',
      'Population age distribution(0-14/60+ years old, %)',
      'International migrant stock(000/% of total pop.)',
      'Education: Primary gross enrol. ratio(f/m per 100 pop.)',
      'Education: Secondary gross enrol. ratio(f/m per 100 pop.)',
      'Education: Tertiary gross enrol. ratio(f/m per 100 pop.)',
      'CO2 emission estimates(million tons/tons per capita)']

for c in Indt.columns:
    if c not in ls:
        temp = [str(v).replace(" ","") for v in Indt.loc[:,c]]
        Indt.loc[:,c] = pd.to_numeric(temp,errors="coerce")
```
